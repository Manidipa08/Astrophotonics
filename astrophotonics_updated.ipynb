# Part 2: Building the complex response matrix of the device

```python
#first making a 2D Energy matrix for each input
#Input0
E_p0_i0 = data1[6,3]
E_p1_i0 = data1[6,4]
E_p2_i0 = data1[6,5]
E_p3_i0 = data1[6,6]

#Input1
E_p0_i1 = data2[6,3]
E_p1_i1 = data2[6,4]
E_p2_i1 = data2[6,5]
E_p3_i1 = data2[6,6]

#Input2
E_p0_i2 = data3[6,3]
E_p1_i2 = data3[6,4]
E_p2_i2 = data3[6,5]
E_p3_i2 = data3[6,6]

#Input3
E_p0_i3 = data4[6,3]
E_p1_i3 = data4[6,4]
E_p2_i3 = data4[6,5]
E_p3_i3 = data4[6,6]


#Making array for each input
E_i0 = np.array([E_p0_i0, E_p1_i0, E_p2_i0, E_p3_i0])
E_i1 = np.array([E_p0_i1, E_p1_i1, E_p2_i1, E_p3_i1])
E_i2 = np.array([E_p0_i2, E_p1_i2, E_p2_i2, E_p3_i2])
E_i3 = np.array([E_p0_i3, E_p1_i3, E_p2_i3, E_p3_i3])

#the 2D energy matrix
E_i = np.array([E_i0, E_i1, E_i2, E_i3])
print("The energy matrix :")
print(E_i)
```

```python
#Making a 2D dphase matrix for the inputs corresponds to 0 output(in degree)
#Input0
dphase_p0_i0 = data1[6,37]
dphase_p1_i0 = data1[6,38]
dphase_p2_i0 = data1[6,39]
dphase_p3_i0 = data1[6,40]

#Input1
dphase_p0_i1 = data2[6,37]
dphase_p1_i1 = data2[6,38]
dphase_p2_i1 = data2[6,39]
dphase_p3_i1 = data2[6,40]

#Input2
dphase_p0_i2 = data3[6,37]
dphase_p1_i2 = data3[6,38]
dphase_p2_i2 = data3[6,39]
dphase_p3_i2 = data3[6,40]

#Input3
dphase_p0_i3 = data4[6,37]
dphase_p1_i3 = data4[6,38]
dphase_p2_i3 = data4[6,39]
dphase_p3_i3 = data4[6,40]

#Making array for each input
dphase_i0=np.array([dphase_p0_i0, dphase_p1_i0, dphase_p2_i0, dphase_p3_i0])
dphase_i1=np.array([dphase_p0_i1, dphase_p1_i1, dphase_p2_i1, dphase_p3_i1])
dphase_i2=np.array([dphase_p0_i2, dphase_p1_i2, dphase_p2_i2, dphase_p3_i2])
dphase_i3=np.array([dphase_p0_i3, dphase_p1_i3, dphase_p2_i3, dphase_p3_i3])

#2D dphase matrix
dphase_i = np.array([dphase_i0*(np.pi/180), dphase_i1*(np.pi/180), dphase_i2*(np.pi/180), dphase_i3*(np.pi/180)])
print("The dphase matrix: ")
print(dphase_i) #dphase in radian
```

Now to get the complex response matrix of the device which is represented by M=A.exp(i$\phi$), we will take sqrt of the Energy matrix to compute A(amplitude) and multiply it with exp(i.dphase matrix).

```python
M = (np.sqrt(E_i))*(np.exp(dphase_i*1j))
print("Complex response matrix: ")
print(M)
```

In the complex matrix, the first column gives positive magnitude values with 0 phase difference. In the case of bright output, there should be consistent phase difference to get constructive interference. and in the first column the imaginary term is zero indicating constructive interference and the sum of the magnitude is higher indicating it is receiving high power. Hence, First column corresponds to the first output is "bright one".  

For the other outputs(say "dark ones"), there are phase variations causing destructive interference, low power. These outputs are not same as the difference in magnitude and phase cause variation in the destructive interferences, leading to different low  power levels at each dark output.

# Perfect Response

We are considering the perfect observing condition:

1. Four identical telescope beams(same amount of light in all)
2. No input phase perturbations

so that means, $ E_{input}=\begin{pmatrix} 1 \\ 1 \\ 1 \\ 1 \end{pmatrix}\;$ is entering into the nuller interferometer.

```python
E_pr_in = np.array([1, 1, 1, 1])
#E_pr_in_T = np.transpose(E_pr_in)
I_pr = (np.abs(np.transpose(M)@E_pr_in))**2 #conventional matrix multiplication

#calculating the I(W/m^2)
#I_pr_T = np.transpose(I_pr)
print("Resulting Intensity in case of perfect response: ")
print(I_pr)

I_pr_sum = I_pr[0]+I_pr[1]+I_pr[2]+I_pr[3]
print("The sum of output intensity: ",np.round(I_pr_sum,2))
```

The first intensity value is significantly higher than the other values indicating it is the brightest one($I_0$).

The best null that the device provides is at the second output(index 1)($I_1$) with the lowest intensity 1.62598803e-04 which corresponds to high degree of nullification of the starlight.

```python
#Making a 2D dphase matrix for the inputs corresponds 0 output(in degree) over the entire wavelength range
#Input0
dphase_p0_i0 = data1[:,37]
dphase_p1_i0 = data1[:,38]
dphase_p2_i0 = data1[:,39]
dphase_p3_i0 = data1[:,40]

#Input1
dphase_p0_i1 = data2[:,37]
dphase_p1_i1 = data2[:,38]
dphase_p2_i1 = data2[:,39]
dphase_p3_i1 = data2[:,40]

#Input2
dphase_p0_i2 = data3[:,37]
dphase_p1_i2 = data3[:,38]
dphase_p2_i2 = data3[:,39]
dphase_p3_i2 = data3[:,40]

#Input3
dphase_p0_i3 = data4[:,37]
dphase_p1_i3 = data4[:,38]
dphase_p2_i3 = data4[:,39]
dphase_p3_i3 = data4[:,40]

#Making array for each input
dphase_i0=np.array([dphase_p0_i0, dphase_p1_i0, dphase_p2_i0, dphase_p3_i0])
dphase_i1=np.array([dphase_p0_i1, dphase_p1_i1, dphase_p2_i1, dphase_p3_i1])
dphase_i2=np.array([dphase_p0_i2, dphase_p1_i2, dphase_p2_i2, dphase_p3_i2])
dphase_i3=np.array([dphase_p0_i3, dphase_p1_i3, dphase_p2_i3, dphase_p3_i3])

#2D dphase matrix
dphase_sum_i = np.array([dphase_i0*(np.pi/180), dphase_i1*(np.pi/180), dphase_i2*(np.pi/180), dphase_i3*(np.pi/180)])
print("The dphase matrix: ")
print(dphase_sum_i) #dphase in radian
```

```python
np.shape(dphase_sum_i)
```

```python
M_wl = (np.sqrt(E_sum_i))*(np.exp(dphase_sum_i*1j))
print("Complex response matrix: ")
print(M_wl)
```

```python
#Again considering the perfect response to get the Intensity for all the wavelengths
for i in range(11):
    #I_wl[i,:] = (np.abs(np.transpose(M_wl[:,:,i])@E_pr_in))**2
    I_wl[i,:] = (np.abs(np.dot(np.transpose(M_wl[:,:,i]), E_pr_in)))**2
    #I_wl_T[:,i] = np.transpose(I_wl[i,:])
    print(f"Intensity over entire bandpass (wavelength {i}) in case of perfect response: ")
    print(I_wl[i,:])


#calculating the I_sum
#I_wl_T = np.transpose(I_wl)
#print("Intensity over entire bandpass in case of perfect response: ")
#print(I_wl_T)
I_wl_result =[]
for i in range(11):
    I_wl_result.append(np.sum(I_wl[i]))

print("The final Intensity over the broad band: ")
q = np.mean(I_wl_result)
print(np.round(q,2))





#print(np.sum(I_pr_sum))
#print("The sum of output intensity: ",np.round(I_pr_sum,2))
```

we can see the degradation in the device performance while considering broad-band. (Here we are considering the perfect response case only). When we are working with design wavelength 1.55 $\mu m$, the total intensity at the output is 3.88 W/$m^2$ but when considering the entire bandpass, the total intensity is reduced to be 3.67 W/$m^2$.

At different wavelength, there might be incoherence in light which reduces effective nulling of the device.

This device is built to achieve higher nulling efficiency at a specific wavelength( in this case, 1.55 $\mu m$) but when we are using the entire spectral band, there is a variation in phase shift over the entire band which makes it difficult to maintain precise destructive interference for nulling (at the dark outputs) and averaging that across different wavelength leads to degradation of the nulling performance, non-effective separation of the bright output from the dark ones, lower intensity at the output.
